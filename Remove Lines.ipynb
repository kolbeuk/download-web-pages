```python
import os

def remove_lines(file_path, lines_to_remove):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with open(file_path, 'w') as file:
        for line in lines:
            if line.strip() not in lines_to_remove:
                file.write(line)

def process_files(substrings_to_remove):
    for filename in os.listdir():
        if filename.endswith('.txt') and os.path.isfile(filename):
            print(f"Processing {filename}")
            remove_lines(filename, substrings_to_remove)
            print(f"Lines removed from {filename}")

if __name__ == "__main__":
    substrings_to_remove = [
        "David Kolb",
        "Recent Posts",
        "See All",
        "  © 2023 David Kolb Consultancy | All rights reserved | ",
        "bottom of page"  # Add your additional strings here
    ]
    
    process_files(substrings_to_remove)
    print("Lines removed from all files.")
```
