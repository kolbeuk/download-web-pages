```python
import os

def remove_lines(file_path, start_substring):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    with open(file_path, 'w') as file:
        for line in lines:
            if start_substring in line:
                file.write(line)
                file.write('"')
                break  # Stop writing after finding the start substring
            else:
                file.write(line)

def process_files(start_substring):
    for filename in os.listdir():
        if filename.endswith('.txt') and os.path.isfile(filename):
            print(f"Processing {filename}")
            remove_lines(filename, start_substring)
            print(f"Content removed from {filename}")

if __name__ == "__main__":
    start_substring = "Learn More about our boutique consulting firm"
    
    process_files(start_substring)
    print("Content removed from all .txt files.")
```
